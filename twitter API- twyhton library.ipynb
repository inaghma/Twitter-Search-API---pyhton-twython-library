```python
#creating a json file using credentials for authentication
import json


credentials = {}
credentials['CONSUMER_KEY'] = '______________'
credentials['CONSUMER_SECRET'] ='_____________________________'
credentials['ACCESS_TOKEN'] = '______________________________-'
credentials['ACCESS_SECRET'] =' ____________________________----'


with open("twitter_credentials.json", "w") as file:
    json.dump(credentials, file)
```

```python
#using the json fiel and twython library
from twython import Twython
import json


with open("twitter_credentials.json", "r") as file:
    creds = json.load(file)

#authentication
python_tweets = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])

#query generation
query = {'q': 'nice',
        'result_type': 'popular',
        'count': 10,
        'lang': 'en',
        }
```

```python
#usign pandas for extracting the necessary columns or dimensions 
import pandas as pd


dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': []}
for status in python_tweets.search(**query)['statuses']:
    dict_['user'].append(status['user']['screen_name'])
    dict_['date'].append(status['created_at'])
    dict_['text'].append(status['text'])
    dict_['favorite_count'].append(status['favorite_count'])


df = pd.DataFrame(dict_)
df.sort_values(by='favorite_count', inplace=True, ascending=False)
df.head(10)
```
